```python
# Importing all the Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',None)
```

```python
# Lets see what is in the Data
df=pd.read_csv('Cleaned.csv')
df.head()
```

```python
# Size of the data 
df.shape
```

```python
df.info()
```

```python
# Stastical Description of Data
df.describe()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(df['Airline'], palette='Set3')
plt.title('Count of Airlines', size=30)
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxenplot(x='Airline', y='Price', data=df, palette='Set3')
plt.title('Airlines vs Price', size=30)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(df['Source'], palette='Set2')
plt.title('Count of Source', size=30)
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxenplot(x='Source', y='Price',data=df, palette='Set3')
plt.title('Airlines vs Source', size=30)
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Day', y='Price',data=df, palette='Set2')
plt.title('Days vs Price', size=30)
plt.xticks(rotation=0)
plt.show()
```

```python
df['Month'] = df['Month'].map({
    1:'JAN',
    2:'FEB',
    3:'MAR',
    4:'APR',
    5:'MAY',
    6:'JUN',
    7:'JUL',
    8:'AUG',
    9:'SEP',
    10:'OCT',
    11:'NOV',
    12:'DEC'
})
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Month', y='Price',data=df, palette='Set2')
plt.title('Month vs Price', size=30)
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Total_Stops', y='Price',data=df, palette='Set2')
plt.title('Stops vs Price', size=30)
plt.xticks(rotation=0)
plt.show()
```
